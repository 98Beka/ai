```python
import numpy as np
import matplotlib.pyplot as plt

# Создаём изображение 10x10 с белым крестом
image = np.zeros((10, 10))  # Чёрный фон (0)
image[4:6, :] = 255  # Горизонтальная линия креста
image[:, 4:6] = 255  # Вертикальная линия креста

# Фильтр размытия (3x3 усредняющий)
filter_blur = np.ones((3, 3)) / 9

# Создаём пустой массив для размытого изображения
blurred_image = np.zeros_like(image, dtype=float)

# Применяем фильтр размытия вручную
for i in range(1, image.shape[0] - 1):
    for j in range(1, image.shape[1] - 1):
        # Извлекаем блок 3x3 вокруг текущего пикселя
        block = image[i-1:i+2, j-1:j+2]
        # Рассчитываем новое значение центрального пикселя
        print()
        print(block)
        print(block * filter_blur)
        print(f"sum: {np.sum(block * filter_blur)}")
        print()
        blurred_image[i, j] = np.sum(block * filter_blur)

# Визуализация
plt.figure(figsize=(15, 5))

# Исходное изображение
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray', interpolation='nearest')
plt.title("Исходное изображение (Крест 10x10)")
plt.colorbar(label="Яркость")
plt.axis("off")

# Размытое изображение
plt.subplot(1, 3, 2)
plt.imshow(blurred_image, cmap='gray', interpolation='nearest')
plt.title("Размытое изображение")
plt.colorbar(label="Яркость")
plt.axis("off")
```
