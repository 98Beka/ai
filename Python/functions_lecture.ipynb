Напишем функцию, которая не получает на вход ни одного аргумента, но запрашивает числа с клавиатуры через пробел. Функция должна сформировать список только из четных чисел.

````python
def get_even():
  str_num = input()
  temp_list = str_num.split()
  res_list = []
  for i in temp_list:
    if int(i)%2 == 0:
      res_list.append(int(i))
  return res_list
````

# Local and Global

В Python существуют два основных типа переменных: локальные и глобальные.

## Локальные переменные:

* Локальные переменные объявляются и используются только внутри определенной функции или блока кода. Они видимы только внутри этой области видимости (scope).
* Локальные переменные имеют более высший приоритет, чем глобальные переменные с тем же именем. Если внутри функции объявлена локальная переменная с определенным именем, то при обращении к этому имени внутри функции будет использоваться локальная переменная, а не глобальная.
Локальные переменные создаются, когда им присваивается значение внутри функции, и уничтожаются при завершении функции.

````python
x = 'Akylai'
def my_function():
    x = 10  # Локальная переменная
    print(x)

my_function()  # Выводит 10
print(x)  # Выводит глобальное значение 'Akylai'
````

## Глобальные переменные:

* Глобальные переменные объявляются вне функций и блоков кода и могут быть использованы в любой части программы.
* Глобальные переменные видимы во всей программе, но их изменение внутри функции требует использования ключевого слова global.
* Глобальные переменные создаются при присваивании значения за пределами функций и блоков кода и остаются в памяти на протяжении всей программы.

````python
x = 10  # Глобальная переменная

def my_function():
    global x  # Используем ключевое слово global, чтобы указать, что мы хотим изменить глобальную переменную
    x = 20
    print(x)

my_function()  # Выводит 20
print(x)  # Теперь x изменилась на 20 и будет выведено 20
````

## Примечание для цикла for

Если переменная создается внутри цикла (например, for или while), она обычно считается локальной переменной, принадлежащей блоку кода цикла и видимой только внутри этого блока кода цикла. Это означает, что эта переменная не будет доступна за пределами цикла, и ее область видимости ограничивается этим циклом.

````python
for i in range(5):
  t = i  # t - локальная переменная, видимая только внутри цикла

print(t)  # Вызовет ошибку, так как t не определена за пределами цикла
````

При использовании стандартного интерпретатора Python переменные, объявленные внутри циклов, считаются локальными по умолчанию и не будут доступны за пределами блока кода цикла.

Разумеется, map стоит применять к итерируемым объектам, таким как список.

````python
a = ['a', 'd', 'g', 'r', 's']
````

````python
def counter(cur=0):

  def step():
    cur += 1
    return cur

  def reset():
    cur = 0
    return cur

  return step, reset
````

````python
a = counter(5)
````

Ошибка **UnboundLocalError: local variable `cur` referenced before assignment** указывает на то, что переменная cur используется внутри функции `step`, но не была объявлена как не локальная переменная.

Если вы хотите изменить переменную, объявленную вне функции, внутри функции, вам нужно использовать ключевое слово `nonlocal` (если переменная объявлена в той же области видимости, но не является глобальной) или `global` (если переменная глобальная).

````python
def counter(cur=0):

  def step():
    nonlocal cur
    cur += 1
    return cur

  def reset():
    nonlocal cur
    cur = 0
    return cur

  return step, reset
````

````python
counter_1 = counter()
counter_2 = counter(2)
````

````python
counter_1[0](), counter_2[0]()
````

# Декораторы Функций

Объявим декоратор. Его смысл в том, что он является функцией, которая "дополняет" функционал какой-либо сторонней функциии.

````python
def func_decorator(func):

  def wrapper():
    'Я учу кота делать интересные вещи!'

    print('---- *кот смотрит на тебя ---')

    func()

    print('--- *кот делает шкодливый вид и убегает ---')

  return wrapper
````

Объявим стороннюю функцию.

````python
def some_cat():
  print('--- Привет! Меня зовут Барсик и я тебе сделал кусь! ---')
````

````python
some_cat()
````

````python
func_decorator(some_cat)
````

````python
cat_ = func_decorator(some_cat)
cat_()
````

````python
@func_decorator
def some_cat():
  print('--- Привет! Меня зовут Барсик и я тебе сделал кусь! ---')
````

Возвратная функция `wrapper()`

````python
def func_decorator(func):

  def wrapper(*args):
    a = '---- *кот смотрит на тебя ---'

    res = func(*args)

    b = '--- *кот делает шкодливый вид и убегает ---'

    return a+res+b

  return wrapper
````

````python
@func_decorator
def some_cat(name:str)->str:

  return f'--- Привет! Меня зовут {name.title()} и я тебе сделал кусь! ---'
````
