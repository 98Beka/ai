# Dict comprehension.



```
{key: value for key, value in f({}) if {}}
```



````python
dict_6 = {'ID_' + str(id): name for id, name in zip(a, b)}
````

````python
dict_6
````

````python
dict_7 = {id: name for id, name in zip(a, b) if name != 'Denis'}
````

````python
dict_7
````

# JSON File

````python
import json
````

## Creating a JSON file from a dictionary

````python
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "skills": ["Python", "Machine Learning", "Data Science"]
}

# Creating a JSON file
with open('data.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
````

## Reading a JSON file into a dictionary

````python
with open('data.json', 'r') as json_file:
    data_loaded = json.load(json_file)

print(data_loaded, f'\n{type(data_loaded)}')
````
