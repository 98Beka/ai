```
my_dict_2 = {}
```

-- это были вполне успешные попытки создать пустой словарь

````python
my_dict_2 = {} # словарь создан
````

````python
type(my_dict_2)
````

## Создав пустой словарь озаботимся его заполнением:



```
dict[new_key] = new_value
```



````python
my_dict_2[2367] = 'Simon'
my_dict_2[7584] = 'Loban'
my_dict_2[8954] = 'Destroyer'
````

````python
my_dict_2
````

## Попробуем создать пару с добавлением повторяющегося ключа:

````python
my_dict_2[8954] = 'Dudkin'
````

````python
my_dict_2
````

Как видим, создания ключа не произошло а произошла перезапись значения по данному ключу. Значение 'Destroyer' сменилось на 'Dudkin'.

Особенность словаря - Все ключи должны быть уникальными. Невозможно создать 2 идентичных ключа. Более того, словарь является неупорядоченным типом данных. Порядок следования пар элементов определить нельзя. У словаря нет индексации.

## Iteration of dict

````python
# Данная процедура итерируется по ключам!
for i in my_dict_2:
  print(my_dict_2[i])
````

## Atributs

Есть отличная возможность достать из словаря отдельно ключи и значения и по ним проитерироваться.

````python
my_dict_2.keys()
````

````python
my_dict_2.values()
````

````python
# iteration by values
for val in my_dict_2.values():
  print(val)
````

````python
# iteration by keys
for key in my_dict_2:
  print(key)
````

````python
# iteration by keys and values
for k, v in zip(my_dict_2, my_dict_2.values()):
  print(k, v)
````

### Попробуем сгенерить словарь!

````python
a = [1213, 2342, 2454, 6535, 5674, 8567, 9876]
b = ['Vasya', 'Masha', 'Kate', 'Iris', 'Petr', 'Denis', 'Karina']
````

````python
dict_4 = {}
for key, val in zip(a, b):
  dict_4[key] = val

dict_4
````

````python
dict_5 = {}
for i, j in enumerate(b):
  dict_5[i] = j

dict_5
````

*enumerate()* позволяет проитерироваться по множеству и предоставить ДВЕ переменные итерации - индекс элемента и сам элемент.

# Dict comprehension.



```
{key: value for key, value in f({множество}) if {условие}}
```



````python
dict_6 = {'ID_' + str(id): name for id, name in zip(a, b)}
````

````python
dict_6
````

````python
dict_7 = {id: name for id, name in zip(a, b) if name != 'Denis'}
````

````python
dict_7
````

### Более злой вариант работы со вложенными списками.

````python
c = [[1213, 'Vasya'], [2342, 'Masha'], [2454, 'Kate'], [5674, 'Petr']]
````

````python
dict_8 = {id: name for [id, name] in c}
````

````python
dict_8
````

# JSON File

````python
import json
````

## Creating a JSON file from a dictionary

````python
data = {
    "name": "Alice",
    "age": 30,
    "city": "New York",
    "skills": ["Python", "Machine Learning", "Data Science"]
}

# Creating a JSON file
with open('data.json', 'w') as json_file:
    json.dump(data, json_file, indent=4)
````

## Reading a JSON file into a dictionary

````python
with open('data.json', 'r') as json_file:
    data_loaded = json.load(json_file)

print(data_loaded, f'\n{type(data_loaded)}')
````
