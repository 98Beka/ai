# Тип данныx list.

Перед нами новый тип данный python! При решении многих задач иногда мы имеем дело с наборами чисел, строк и других единиц информации разных типов. Оперировать ими по отделльности чаще всего нерационально. Хорошим решением является придумать какой либо "Контейнер" или коробку, куда эти самые объекты можно сложить и хранить вместе! Эту задачу решают так называемые "Листы. Лист - это упорядоченный набор данных, кторые могут иметь свой определенный тип. Лист может содержать строки, числа, другие листы, словари итд... Но, каждый элемент листа закреплен на своем определенном месте и имеет индекс(порядковый номер). Лист относится к ИЗМЕНЯЕМОМУ типу данных, что предполагает возможность вносить в него изменения. Строка такой возможности не дает, так как она представляет данные неизменяемого типа! Строку невозможно изменить а только пересоздать заново.

Списки или Листы часто также называют 'Массивами', 'Векторами', 'наборами', 'множествами' и иначе.

Выше пример пустого списка, в котором нет ни одного элемента.

```python
# Примеры непустых множеств (списков)
a = [1, 2, 3, 4, 5, 6, 7]                   # список чисел (гомогенный список. Не путать с гомофобным!)
b = ['a', 'b', 'c', 'd', 'e', 'f']          # список строк
c = [1, 2, 'a']                             # список с данными смешанного типа (гетерогенный список)
d = [[1, 2, 3], ['a', 'd'], [10, 202, 30]]  # список со списками (вложенный список)
```

# list()

Это функция, которая может реализовывать итеративный алгорим, "пробегаясь" по элементам какого-либо множества (строка - это тоже множество) и каждый элемент итерируемого объекта складывать в новообразованный лист!

```python
x = 'Ну да, я строка и что ты мне сделаешь?'
# а вот что:

x_list = list(x)
print(x_list) # каждый элемент строки стал элементом списка.
```

```python
x = range(1, 11)
x # числовое множество от 1 до 10
```

```python
list(x)
```

```python
list(range(1, 20, 2))
```

# Slicing. Correct with slicsing.

Спойлер: работает в точности также как и для строк!!!

Но так как список относится к ИЗМЕНЯЕМЫМ типам данных, то обратившись по индексу к элементу его можно ЗАМЕНИТЬ на новый или выбрать номер ячейки в списке и туда перезаписать новый элемент

```python
a = [1, 2, 3, 4, 5, 6, 7]
print(a[2])
print(a[1:-1])
print(a[::-1])
```

```python
a[1] = 200
a # произошла замена 2 на 200
```

```python
a[2:4] = [30, 40]
a # заменили 2 элемента
```

```python
a[4] = a[4]*1000
# a[4] *= 1000   сокращенный вариант записи
a
```

## append(value)

Данная функция добавляет элемент в конец строки. Изменения вступают в силу сразу!

```python
l = ['a', 'b', 'c']
l.append('d')
l
```

## insert(index, value)

По указанному индексу добавляет элемент. А все элементы справа сдвигает на одну позицию. Изменения вступают в силу сразу!

```python
l.insert(2, 'bbb')
l
```

```python
l.insert(20, 'bbb') # что тут произошло?
print(l)
l.index('bbb')
print(f"lust 'bbb' inedx:{len(l) - 1 - l[::-1].index('bbb')}")
```

## extend(list)

Объединяет 2 списка. Аналог "+"

```python
a = ['a', 'b', 'c', 'd']
b = ['e', 'f', 'g', 'h']
a.extend(b)
a
```

## pop(index)

"Поп достал из старинного сундука книгу и дал мне..."

Извлекает из списка элемент по указанному индексу и возвращает его.

```python
x = ['a', 'b', 'c', 'd']
x_0 = x.pop(0)
print(x)
x_0
```

```python
x # 'a' больше в списке нет
```

## remove(value)

Удалаяет из списка указанный элемент. Если таких элементов несколько, то удаляет тот, который встречается в списке первый раз.



```python
x = [1, 2, 3, 4, 5]
x.remove(4)
x
```

```python
y = ['a', 'b', 'c', 'd']
y.remove('b')
y
```

```python
z = ['a', 'b', 'c', 'd', 'b', 'b', 'b']
z.remove('b')
z
```

## index(value, start, stop)

Возвращает индекс указанного элемента. Если их несколько, то индекс первого вхождения элемента в список. Может работать в указанном диапазоне значений индексов.

```python
z = ['a', 'b', 'c', 'd', 'b', 'b', 'b']
z.index('b')
```

```python
z.index('b', 2)
```

```python
# задача найти и заменить
x = [1, 2, 3, 4, '5', 6, 7]
x[x.index('5')] = 5
x
```

## sort()

Выполняет сортировку списка. Может сортировать как по возрастанию, так и по убыванию. Может сортировать по алфавиту.

```python
x = [3, 2, 5, 4, 6, 5, 7, 10, 0]
x
```

```python
x.sort()
x
```

```python
x.sort(reverse=True)
x
```

```python
y = ['f', 'a', 'g', 'b', 'd', 'c']
y.sort()
y
```

```python
y.sort(reverse=True)
y
```
