## Задача. Отберем из листа только уникальные элементы

````python
list_ = [1, 2, 3, 4, 5, 3, 5, 4, 2, 6, 5, 7, 8, 7, 10, 3, 10]
````

````python
unique_list = []
for i in list_:
  if i not in unique_list:
    unique_list.append(i)

unique_list
````

# list comprehension

Это вариант генерации списка с использованием цикла но в более лаконичной форме. Такая форма делает код более коротким и читабельным. Использование list comprehension является индикатором профессионализма. Вообще, в профессиональном коде использование цикла for является не совсем приличным, если без него обойтись нельзя.

Общий вид синтаксиса:



```
new_list = [f(i) for i in {итерируемое множество}]

new_list = [f(i) for i in {итерируемое множество} if {условие==True}]

new_list = [f(i) if {условие==True}  else g(i) for i in {итерируемое множество}]
```



$f(i) = const$

````python
list_ = ['a' for _ in range(10)]
list_
````

$f(i) = i$

````python
comp = [i**2 + 2 for i in range(10)]
comp
````

````python
cop = ['asd','zxc','fgh','vac']
aaa = [i.title() for i in cop if 'a' in i]
aaa
````

````python
cop = ['asd','zxc','fgh','vac']
aaa = [ i.title() if 'a' in  i  else None for i in cop ]
aaa
````

$f(i) = i^2$

````python
sq_list = [i**2 for i in range(5)]
sq_list
````

## Задача.
Найдем целые корни квадратного уравнения итеративном методом.

$$x^2 -12x + 32 = 0$$

````python
# классический метод
%%time

roots = []
for x in range(-100, 100):
  if x**2 - 12*x + 32 == 0:
    roots.append(x)

roots
````

````python
# list comprehension
%%time

roots = [x for x in range(-100, 100) if x**2 - 12*x + 32 == 0]
roots
````

Тут мы замерили время выполнения ячеек и увидели, что *list comprehension* работает быстрее!

## Задача.

Даны два числовых списка с замерами температур воздуха. Первый список - утром, второй - вечером. Для каждой пары отберите минимальное значение температуры и сложите их в новый список.

````python
morning = [12, 15, 21, 24, 17, 19, 20, 10]
evening = [15, 20, 22, 15, 16, 20, 13, 5]
````

````python
dlt = [min(m, e) for m, e in zip(morning, evening)]
dlt
````

````python
[[m, e] for m, e in zip(morning, evening)]
````
