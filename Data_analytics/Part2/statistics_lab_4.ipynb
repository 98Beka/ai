# Task_5

Изучите способ создания специфических кластеров.
Построим множество точек в количестве 500, которые распределены вдоль прямой

$$y = 2x-1$$

```python
x = np.linspace(-10, 10, 500)
y = 2*x - 1
```

```python
sns.scatterplot(x=x, y=y)
```

Добавим "Гаусовый шум" - случайное отклонение от прямой для каждой точки

```python
y = 2*x - 1 + np.random.randn(500)
```

```python
sns.scatterplot(x=x, y=y)
```

Гаусовый шум можно масштабировать добавив множитель:

```python
y = 2*x - 1 + np.random.randn(500)*5 # множитель "5"
sns.scatterplot(x=x, y=y)
```

Также множество $X$ тоже можно задать случайным:

```python
x = np.random.randn(500)
y = 2*x - 1 + np.random.randn(500)
sns.scatterplot(x=x, y=y)
```
