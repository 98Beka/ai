Расчитаем произведение всех отклонений по двум переменным. Большая часть этих отклонений будет положительным числом.


\begin{cases}
    (x_1 - \bar{x})(y_1 - \bar{y}) > 0 \\
    (x_2 - \bar{x})(y_2 - \bar{y}) < 0 \\
    . \\
    . \\
    . \\
    . \\
    (x_n - \bar{x})(y_n - \bar{y}) ... 0 \\
  \end{cases}



---

Сложим все эти отклонения и усредним. Мы получим показатель, который называется ковариацией.

$$cov = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}{N-1}$$

Его смысл в том, что он показывает СИЛУ и НАПРАВЛЕНИЕ взаимосвязи двух переменных.

image.png

Так как сами значения переменных могут быть сколь угодно большими и имеют единицы измерения, значение $cov$ порой сложно интерпретировать. Хорошо было бы этот показатель привести к какой либо безразмерной величине с масштабировать. Для этого применяют нормирование $cov$ по стандартным отклонениям для каждой переменной. Тогда мы получим безразмерную величину в интервале $[-1; 1]$.

$$ r_{xy} = \frac{cov}{\sigma_x \sigma _y}$$  $$r_{xy} \in [-1; 1]$$

Данный показатель получил название Коэффициент корреляции Пирсона.


```python
def corr_x(arr):
  arr_mean = arr.mean(axis=0)
  cov = ((arr[:, 0] - arr_mean[0]) * (arr[:, 1] - arr_mean[1])).sum()/(arr.shape[0] - 1)
  r_xy = cov/(arr[:, 0].std()* arr[:, 1].std())
  return r_xy
```

Если коэффициент кор Пирсона показывает силу и направление взаимосвязи, то коэф детерминации показывает часть изменчивости одной переменной, которая обусловлена тем, что она "взаимосвязана" с другой переменной. Если коэффициент корреляции значимо отклоняется от нуля, то в ощутимой мере мы говорим, что одна переменная изменяется не случайно, а в некотором соответствии с тем, как изменяется другая переменная.

# Проверка гипотезы

Мы можем применять в статистическом анализе скоррелированность данных используя t-критерий Стьюдента.

$H_0: r_{xy} = 0 $

$H_1: r_{xy} \neq 0 $

Нулевая гипотеза утверждает что в ГС нет никакой корреляции между исследуемыми признаками. Альтернативная - наоборот.

Для получения t-критерия необходимо получить число степеней свободы:

$$df = N-2$$

-2 так как у нас теперь две переменных.

Для расчета существует достаточно интуитивно непонятная формула, но примем к сведению, что она есть:

$$ t = |r_{xy}| \sqrt {\frac{df}{1-r_{xy}^2}}$$

https://www.socscistatistics.com/pvalues/tdistribution.aspx
