## Getting the data

We'll use two .csv files for our data. The first contains information on life expectancy and the second has the data on the population breakdown by age and sex.

Columns in **UKCountriesLifeExpectancy.csv**:
- *Year*
- *Var*: Life expectancy at birth or Life expectancy at 65
- *Country*: England, Scotland, Wales and N Ireland
- *N*	- Life expectancy
- *Sex*



Columns in **UKRegionsPopulation.csv**:
- *Country*: England, Scotland, Wales and N Ireland
- *Children*
- *Adults*
- *Pensioners*
- *Male*
- *Female*

```python
demographic = pd.read_csv("UKRegionsPopulation.csv", index_col = 0)
demographic_child = pd.DataFrame(demographic.Country)
demographic_child['count'] = demographic.Children
demographic_child['who'] = 'Children'
demographic_adults = pd.DataFrame(demographic.Country)
demographic_adults['count'] = demographic.Adults
demographic_adults['who'] = 'Adults'
demographic_pensioners = pd.DataFrame(demographic.Country)
demographic_pensioners['count'] = demographic.Pensioners
demographic_pensioners['who'] = 'Pensioners'
demographic_merg = pd.merge(demographic_child, demographic_adults, how='outer').merge(demographic_pensioners, how='outer')
demographic_sex_male = pd.DataFrame(demographic.Country)
demographic_sex_male['count'] = demographic.Male
demographic_sex_male['sex'] = 'Male'
demographic_sex_female = pd.DataFrame(demographic.Country)
demographic_sex_female['count'] = demographic.Female
demographic_sex_female['sex'] = 'Female'
demographic_sex = pd.merge(demographic_sex_male, demographic_sex_female, how='outer')

demographic_merg
```
