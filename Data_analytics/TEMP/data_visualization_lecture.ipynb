# Matplotlib
Python library that allows us to visualize numeric data as plots and graphs.

Makes easy things easy (plotting basic graphs) and hard things possible (multiple graphs, subplots)

````
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Magic line that allows us to see plots inline. No command = file output
%matplotlib inline
````

## Plotting basics

````
y_values = [1, 2, 3, 4]
plt.plot(y_values, color='green')
````

A wild graph appeared! What happened?
* Y-axis values are the values from the passed in array
* X-axis values are incrementing from index 0 to N-1
* A straight line is drawn through each point

````
# Plot points instead of a line
plt.plot(y_values, 'xb')
````

The second argument takes a String for how to draw the graph. The string uses the following format:

`fmt = '[marker][line][color]'`

Examples include:

`'ob' # Blue circle markers`

`'-g' # Green line`

For more on the format, see:
https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html

````
x_data = [1, 2, 3, 4]
y_data = [1, 4, 9, 16]
plt.plot(x_data, y_data)
````

We can also use numpy to generate arrays for plotting points. In the next example, we'll use the `linspace` API to generate an array of linearly increasing points:

````
linear_data = np.linspace(0, 10, 25) # (start float value, end float value, # of linear points)
print(*linear_data, sep='\n')
````

````
plt.plot(linear_data, linear_data, 'og')
````

````
plt.plot(linear_data, linear_data, '.r')
plt.title("Simple red plot using linspace")
````

````
plt.plot(linear_data, linear_data, '.m', label='plot')
plt.title("Simple Magenta plot using linspace")
plt.xlabel("x label")
plt.ylabel("y label")
plt.legend()
````

## Plotting multiple graphs
We can plot multiple graphs on one figure or have multiple subplots on one figure

````
linear_data = np.linspace(0, 10, 125)
````

````
# plt.plot(linear_data, np.tan(linear_data), label="tg")
plt.plot(linear_data, np.cos(linear_data), label="cos")
plt.plot(linear_data, np.sin(linear_data), label="sin")
plt.legend()
plt.title("Simple Plot with 3 lines")
plt.xlabel("x label")
plt.ylabel("y label")
````

$y = \frac{1}{2}x^2 - 3x +6$

````
x = np.linspace(-15, 20, 100)
plt.plot(x, .5*x**2 -3*x +6)
````

````
# nrows, ncols
# figsize = (width, height) in inches
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

ax[0].plot(linear_data, linear_data, '+b', label="Linear")
ax[1].plot(linear_data, linear_data**2, '.g', label="Quadratic")
ax[2].plot(linear_data, linear_data**3, 'or', label="Cubic")
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.suptitle("Simple plot with 3 subplots")

plt.xlabel("x label")
plt.ylabel("y label")
````

## Bar graphs
Suppose we had a dataset of test scores for a given class. How could we get a general sense of how well the class did on the test overall? Given that bar graphs are good for:

- Comparing different variables together
- Seeing general trends of the data

Let's use a bar graph!

First, let's set up the dataset of test scores for students based on a normal distribution with a mean of 70 and a standard deviation of 5:

![alt text](https://www.syncfusion.com/books/Statistics_Using_Excel_Succinctly/Images/normal-curve.png)

````
# Generate data
num_students = 300
mean = 70
std = 5
data = np.random.normal(mean, std, num_students)

data[:10]
````

Let's segment out the data into the following segments:
 - How many people got below 65%?
 - How many people got between 65% and 75%?
 - How many people got above 75%?

````
below_65 = (data < 65.0).sum()
between_65_75 = ((data >= 65.0) & (data <= 75.0)).sum()
above_75 = (data > 75.0).sum()

score_data = [below_65, between_65_75, above_75]
print("Score data: ", score_data)

x=np.arange(1, 4)
plt.title("Test Score Distribution")
plt.xlabel("Score")
plt.ylabel("# of students")
plt.xticks(x, ("< 65%", "65% - 75%", ">75%"))
plt.bar(height=score_data, x=x)
````

# Seaborn
* Seaborn is a convenience layer library on top of Matplotlib
* It aims to make a certain subset of hard things in matplotlib easy
 * Visualizing data from Pandas objects such as matrices, dataframes (Pandas was created more than a decade after matplotlib)
 * Enhancing color and theming options by default
 * Automatic estimation and statistical analysis built-in
* Seaborn is designed to augment matplotlib, not replace it

````
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns # Shorthand
````

Let's plot some basic data analysis using Seaborn. We can easily create a scatter plot comparing Pokemon's attack vs. defense ratings.

## Colors
Let's take a quick detour and talk about colors in a graph. Colors can greatly help in better conveying information via your data visualizations. Colors can be used to:

* Highlight specific aspects of your data
* Differentiate between categories of data
* Indicate changes in your data

Let's use color in our scatterplot above to take our visualization to the next level. Seaborn allows us to specify a 'hue' attribute that will color the points based on a separate column as a 'Stage' or 'Type 1':

````
# Look at the Stage block, there's 4 stages, stage 3

sns.scatterplot(x='Attack', y='Defense', data=data, hue='Stage')
````

A wild legend has appeared!

Let's visualize our Pokemon's Attack vs. Defense stats based on their type:

````
sns.scatterplot(x='Attack', y='Defense', data=data, hue='Type 1')
````

We can even customize the color's palette to match the Pokemon's type color in Bulbapedia:

````
pkmn_type_colors = {'Grass': '#78C850',
                    'Fire': '#F08030',
                    'Water': '#6890F0',
                    'Bug': '#A8B820',
                    'Normal': '#A8A878',
                    'Poison': '#A040A0',
                    'Electric': '#F8D030',
                    'Ground': '#E0C068',
                    'Fairy': '#EE99AC',
                    'Fighting': '#C03028',
                    'Psychic': '#F85888',
                    'Rock': '#B8A038',
                    'Ghost': '#705898',
                    'Ice': '#98D8D8',
                    'Dragon': '#7038F8',
                   }
sns.scatterplot(x='Attack', y='Defense', data=data, hue='Type 1', palette=pkmn_type_colors)
````

## Histograms in Seaborn

Let's view a few histograms in order to see bucketing trends for various columns

````
sns.distplot(data['Attack'], bins=15)
````

Seaborn will also include a line graph in addition to the bar graph for the bins. This line graph is what's known as the Kernel Density Estimate (KDE) and provides a general 'shape' of the data as a line graph. It is useful for visualizing the discrete bins as a continuous line.

To learn more about KDE, see here: https://mathisonian.github.io/kde/

````
sns.distplot(data['HP'], bins=20)
````

We can visualize multiple histograms on one plot:

````
f, axes = plt.subplots(2, 2, sharex=True)
sns.distplot(data['HP'] , color='green', ax=axes[0, 0]).set_title('green')
sns.distplot(data['Speed'] , color='yellow', ax=axes[0, 1]).set_title('yellow')

sns.distplot(data['Attack'] , color='red', ax=axes[1, 0])
sns.distplot(data['Defense'] , color='blue', ax=axes[1, 1])
````

## Correlation heatmap in Seaborn
One special type of plot we can use is a correlation heatmap that will allow us to see if any columns are correlated to each other. The values in a correlation heatmap range from [-1, 1], 1 meaning it is completely correlated, 0 meaning there is no correlation, and -1 meaning it's completely inversely correlated.

````
corr = data.corr()

# Heatmap
sns.heatmap(corr)

# Size vs. speed
# Why is the diagonal always one?
# Why are the values mirrored diagonally?
````

Let's add annotations to our heatmap to more accurately see if there are any correlations.

````
corr
````

````
sns.heatmap(corr, annot=True, fmt='0.2f')
````

## Challenge

Now that we've learned a lot about Seaborn and how to use visualizations to learn more about our data, let's try to solve this problem:

What is the strongest Pokemon type on average, strongest being defined as a high Attack and high Defense rating? How would we go about visualizing this? Discuss with your group.


---

One way to tackle this problem is to find the average Attack and Defense ratings for each Pokemon type and compare them against each other.


![alt text](https://i.pinimg.com/originals/20/df/07/20df0752b38ec109f2f39cd9fe88bf98.png)

````
avg_type_data = data.groupby(['Type 1']).mean()
avg_type_data.head()
````

````
# Let's make sure the Type 1 is also a column
avg_type_data.reset_index(level=0, inplace=True)
avg_type_data.head()
````

````
avg_type_data
````

````
avg_attack_type_data = avg_type_data.sort_values(by=['Attack'])
sns.barplot(x=avg_attack_type_data['Type 1'], y=avg_attack_type_data['Attack'], palette=pkmn_type_colors)
````

````
avg_defense_type_data = avg_type_data.sort_values(by=['Defense'])
sns.barplot(x=avg_attack_type_data['Type 1'], y=avg_defense_type_data['Defense'], palette=pkmn_type_colors)
````

Let's use a scatterplot to visualize the average attack + defense ratings together:

````
sns.scatterplot(x='Attack', y='Defense', data=avg_type_data, hue='Type 1', palette=pkmn_type_colors)
````

Hmm the dots look a little small. Let's increase their size by using the `s` parameter:

````
sns.scatterplot(x='Attack', y='Defense', data=avg_type_data, hue='Type 1', palette=pkmn_type_colors, s=1000) # s units are in 'points'
````

Let's try adding some text for each point. We can do this by using matplotlib's `text` API:

````
plot = sns.scatterplot(x='Attack', y='Defense', data=avg_type_data, hue='Type 1', palette=pkmn_type_colors, s=2000)
for pkmn_type in range(0, avg_type_data.shape[0]):
  plot.text(
      avg_type_data['Attack'][pkmn_type],
      avg_type_data['Defense'][pkmn_type],
      avg_type_data['Type 1'][pkmn_type],
      horizontalalignment='center',
      size='small',
      color='white',
      weight='semibold'
  )
````
