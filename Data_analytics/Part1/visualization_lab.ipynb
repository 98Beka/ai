# Task_6

Подумайте, как можно использовав множества $x$ и $y$, полученные в предыдущей задаче, чтобы построить окружность со смещением ее центра в точку с заданными координатами $(x; y)$.

```python
def get_circle(r, x, y):
  t = np.linspace(0, 2*np.pi, 1000)
  arr_x = r * np.cos(t) + x
  arr_y = r * np.sin(t) + y
  return arr_x, arr_y
```

```python
for i in range(1, 6):
  x, y = get_circle(i, 2+i, 1+i)
  plt.plot(x, y, color='red')
```

# Task_8

Напишите функцию, которая отрисовывает равносторонный треугольник с заданной стороной. Основание треугольника должно занимать горизонтальное положение а центр треугольника в начале координат.

```python
def drow_triangle(lentgh:int):
    h = np.sqrt(lentgh**2 - (lentgh/2)**2)
    a = np.array([[-lentgh/2, -h/2], [0, h/2]])
    b = np.array([[lentgh/2, -h/2], [0, h/2]])
    c = np.array([[-lentgh/2, -h/2], [lentgh/2, -h/2]])
    plt.plot(a[:,0], a[:,1], color='red')  
    plt.plot(b[:,0], b[:,1], color='red')  
    plt.plot(c[:,0], c[:,1], color='red')

drow_triangle(10)
```

# Animation

Ниже написан пример, как можно анимировать некоторые процессы. Анимация представляет собой последовательность кадров и в каждом кадре происходит обновление графика.

В указанном примере в каждом кадре генерится точка из нормального распределения и наносится на scatterplot.

## Params

* `fig` - системо координат заданного размера с заданным диапазоном разметки осей
* `update` - функция обновления графика
* `frames` - количество кадров
* `interval` - временной промежуток между кадрами в [мс]
* `repeat` - повтор.

```python
from IPython.display import display, clear_output
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)

def update(frame_number):
    sns.scatterplot(x=np.random.randn(1), y=np.random.randn(1), color='green')


ani = FuncAnimation(fig, update, frames=100, interval = 50, repeat=False)
HTML(ani.to_html5_video())
```

В данном примере в каждом кадре отрисовывается равномерное распределение точек в количестве 500. В каждой следующей итерации пространство очищается от точек предыдущего кадра.

```python
fig, ax = plt.subplots(figsize=(5, 5))
def update(frame_number):
    # Очистка графика
    ax.cla()
    # отрисовка графика
    ax.set_xlim(-1, 11)
    ax.set_ylim(-1, 11)
    sns.scatterplot(x=np.random.randint(1, 1000, 500)/100, y=np.random.randint(1, 1000, 500)/100, color='green')


ani = FuncAnimation(fig, update, frames=100, interval = 50, repeat=False)
HTML(ani.to_html5_video())
```

# Task_14

Сдедайте анимацию качающегося маятника в виде "кружочка" на нитке. Из параметров, которые можно задать - это частота колебаний. Но вы можете произвольно добавить свои параметры.

```python
DURATION = 7 #sec
FR = .5 # Hz
FPS = 40 # Frame per second
LENGHT = 7 # Длина нити (от подвеса до центра окружности)
RADIUS = 1
AMPLITUDA = 3
```

```python
t = np.linspace(0, DURATION, DURATION*FPS)
t.shape
```

$$x = A*\sin{(2\pi \nu t)}$$

$$y = - \sqrt{l^2 - x^2}$$

```python
def get_coords():
  time_ind = -1

  def frame():
    nonlocal time_ind
    time_ind += 1
    x = AMPLITUDA * np.sin(2*np.pi * FR * t[time_ind])
    y = - (LENGHT**2 - x**2)**.5
    return x, y

  return frame

def get_circle(r, x, y):
  t = np.linspace(0, 2*np.pi, 1000)
  arr_x = r * np.cos(t) + x
  arr_y = r * np.sin(t) + y
  return arr_x, arr_y
```

```python
coords = get_coords()
fig, ax = plt.subplots(figsize=(5, 5))
def update(frame_number):
    # Очистка графика
    ax.cla()
    # отрисовка графика
    ax.set_xlim(-AMPLITUDA - RADIUS -1, AMPLITUDA + RADIUS + 1)
    ax.set_ylim(-LENGHT-RADIUS-1, 1)
    x, y = coords()
    arr_x, arr_y = get_circle(RADIUS, x, y)
    sns.scatterplot(x=[x], y=[y], color='green')
    plt.plot(arr_x, arr_y, color='green')
    plt.plot([0, x], [0, y], color='green', marker='o')



ani = FuncAnimation(fig, update, frames=DURATION*FPS-2, interval = 1000/FPS, repeat=False)
HTML(ani.to_html5_video())
```
