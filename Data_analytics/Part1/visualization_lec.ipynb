# Seaborn

Это одна из самых популярных библиотек для визуализации. Во многом повторяет функционал `matplotlib` и может использоваться самостоятельно. В некоторых случаях, для построения сложных графиков и дашбордов может использоваться совместно с `matplotlib`.

# Lineplot

* `sns.lineplot()`
* `plt.plot()`

Данные функции позволяют строить простейшие графики, подобные тому, какие мы строили в школе. `sns.lineplot()` - метод библиотеки `seaborn`, а `plt.plot()` - библиотеки `matplotlib`. Мы в основном будем использовать `seaborn`.

`sns.lineplot()` получает на вход много настроечных параметров, но главными являются переменные $x$ и $y$, зависимость между которыми мы хотим показать.

````python
x = [1, 2]
y = [3, 2]
````

Создали два массива, где в первом координаты $x$ для двух точек, и во втором их координаты $y$.

`sns.lineplot()` построит две точки

* $(1, 3)$
* $(2, 2)$

и соединит их прямой линией.

````python
sns.lineplot(x=x, y=y)
````

````python
plt.plot(x, y)
````

Покажем более сложный вариант:


````python
x = [1, 2, 3, 4]
y = [1, 4, -1, 6]
````

````python
sns.lineplot(x=x, y=y)
````

Есть опция задать цвет линии и название графика.

````python
sns.lineplot(x=x, y=y, color = 'red').set_title('My First Plot')
````

# Графики функций.

Данные графики строятся на плоскости, поэтому в основном умеют показывать зависимость двух переменных. Для любого графика нам необходимо получить:

* Множество аргумента $X$
* Множество значений функции $Y$

Построим график простейшей функции $$y = x^2$$

````python
x = np.arange(-3, 4)
x # множество аргумента, для каждого элемента которого будем расчитывать значение функции
````

````python
y = x**2
y
````

````python
sns.lineplot(x=x, y=y)
````

# `np.linspace(start, stop, num)`

График получился "угловатым" так как наше множество аргумента имеет мало значений с очень большим шагом. Если мы сделаем точек больше, например 50, то кривая станет значительно более гладкой. Для этого удобным инструментом является функция `np.linspace()`

````python
np.linspace(-3, 3, 50)
````

Мы задали диапазон от $-3$ до $3$ и заполнили этот диапазон числами в количестве $50$ равным шагом. Шаг расчитывается автоматически.

````python
x = np.linspace(-3, 3, 50)
y = x**2
````

````python
sns.lineplot(x=x, y=y, color='green')
````

Построим в одной системе координат два графика

$$y = x^2$$
$$y = \sqrt{x}$$

Для $x \in [0, 4]$

````python
x = np.linspace(0, 4, 100)
y_1 = x**2
y_2 = x**.5
````

````python
sns.lineplot(x=x, y=y_1, color = 'red', label='Квадрат') # label - значение для легенды в графике.
sns.lineplot(x=x, y=y_2, color = 'green', label='Квадратный корень')
````

В таких случаях не всегда удачно подбирается масштаб. На это можно повлиять, если настроить размер окна с графиком.



```
plt.subplots(figsize=(5, 10))
```
Позволяет создать рабочее поле с заданным размеров. Все графики, коорые будут написаны ниже будут "вписаны" в данное окно (фигуру).


````python
plt.subplots(figsize=(3, 10))
sns.lineplot(x=x, y=y_1, color = 'red', label='Квадрат')
sns.lineplot(x=x, y=y_2, color = 'green', label='Квадратный корень')
````

# Scatterplot

## `sns.scatterplot()`

Данный тип графика называется также диаграммой рассеяния. Как и Lineplot он показывает зависимость одной переменной от другой, но особо удобен тогда, когда множество точек невозможно выстроить в линию.

Возмем набор данных и попробуем повизуализировать.

````python
x  = np.random.randn(500)
y  = np.random.randn(500)
x[:10], y[:10]
````

````python
sns.lineplot(x=x, y=y)
````

Как видим, данный график не показывает ничего удобовразумительного. Попробуем scatterplot:

````python
sns.scatterplot(x=x, y=y)
````

Этот график очень полезен для понимания особенностей данных. Позволяет увидеть такие факторы как:
* Характер распределения
* Наличие тренда
* Шумы и выбросы
* Наличие выраженных кластеров

Давайте взглянем на несколько примеров таких диаграмм и оценим различия между ними.

````python
x = np.random.randn(400)
y = x + 2 + np.random.randn(400)*.5
sns.scatterplot(x=x, y=y)
#sns.lineplot(x=np.linspace(-3,3,100), y = np.linspace(-3,3,100)+2, color='red')
````

````python
x = np.random.randn(400)
y = -x + 2 + np.random.randn(400)*.8
sns.scatterplot(x=x, y=y)
#sns.lineplot(x=np.linspace(-3,3,100), y = -np.linspace(-3,3,100)+2, color='red')
````

````python
x = np.random.randn(400)*2
y = x**2 + 2 + np.random.randn(400)*4
sns.scatterplot(x=x, y=y)
#sns.lineplot(x=np.linspace(-6,6,100), y = np.linspace(-6,6,100)**2+2, color='red')
````

# Параметр `hue` и цветовая дифференциация.

Многие графики могут кроме непосредственной зависимости между переменными показывать также другие факторы с помощью цвета, формы маркеров, размера точки или маркера и толщины линий. Часто для scatterplot испоьзуется цветовая маркировка, чтобы показать дополнитеьлные свойства объектов данных. (Каждая точка является объектом)

image.png

Тут мы видим таблицу, где каждая строка является объектом, и у каждого объекта есть 2 признака. Например это может быть таблица с вашими клиентами, о каждом из которых вы знаете возраст и сколько он зарабатывает долларов в час. Для визуализации таких  данных вам очень хорошо подойдет scatterplot, где два признака как раз станут координатами каждой точки.

image.png

Но что делать, если у нас появляется еще один признак, который мы тоже хотим показать на графике? Можно построить трехмерный scatterplot, но это не всегда удобно. Часто прибегают к методике, когда третий признак показывается с помощью цвета.

````python
matrix[:5]
````

Для нашего набора объектов с двумя признаками добавим еще один признак для каждого объекта в виде отдельного массива. (Пока отдельного)

````python
cat = ['A' for _ in range(300)]
cat.extend(['B' for _ in range(300)])
cat = ['A'] * 300 + ['B'] * 300 + ['C'] * 300 + ['D'] * 20
````

Мы создали список с символами 'A' и 'B', в котором количество элементов равно количеству строк в нашей таблице. Таким образом каждому объекту из таблицы по соответствующему индексу будет поставлена в соответствие категория из списка.

````python
sns.scatterplot(x=matrix[:, 0], y=matrix[:, 1], hue=cat)
````

# Clusters & Noise

Взглянем снова на график из предыдущих примеров.

````python
def range_():
  return np.random.randn(300, 2)
matrix = np.concatenate([range_(), range_()+np.array([4, 6]),
                         range_()+np.array([7, -3]),
                         np.random.randn(20, 2)*9], axis=0)

sns.scatterplot(x=matrix[:, 0], y=matrix[:, 1])
````

## Clusters
Мы можем на данном графике очень наглядно видеть то, что большинство объектов формируют ГРУППЫ в силу похожести своих свойств. Это подобно тому, как мы, люди, выбираем круг общения по своим специфическим интересам, и в сколь-либо публичном месте мы можем наблюдать все множество людей собранных по отдельным группам сообразно специфики профессии, интересам, виду деятельности и так далее.
В науке о данных такие группы мы называем КЛАСТЕРАМИ (clusters), и их обнаружение является принципиально важным для анализа и понимания особенностей данных. Принадлежность к какому-либо кластеру является вполне себе ПРИЗНАКОМ. Иногда принадлежность к кластеру предопределена изначально, а иногда это может быть предметом открытия при изучении данных.

## Noise

Но наш мир не идеален и всегда есть особенности и исключения. Есть люди - интроверты, которые склонны к тому, чтобы не быть частью никакой из групп. Они предпочитают держаться в стороне от всех.

В данных также есть объекты, которые могут формировать кластеры, но есть такие, которые не принадлежат ни к какому из кластеров. Такие объекты мы крайне обидно называем Шумовыми или Аутлаерами.
